```python
import pandas as pd
```

```python
'''
Removes sequences of repeating pages from a users session

Parameters:
data - cleaned data frame with user_journeys seperated by '-'

Returns:
A copy of the data frame with concurrent page visits removed
'''
def remove_page_duplicates(data, target_column='user_journey'):
    
    # Creates a new copy of the data frame that is passed as a variable into the function
    cleaned_data = data.copy()
    
    # Iterates through each row of the dataframe
    for index, row in data.iterrows():
        
        # Split the user journey string into a list of seperate strings for words separated by a '-'
        split_list = row[target_column].split('-')
        
        # Initialise a new list to store the cleaned journey, starting with the first string in the list
        result_list = [split_list[0]]
        
        # Iterate through each of the strings in the list from the second value
        for item in split_list[1:]:
            if item != result_list[-1]:
                result_list.append(item)
        
        # Join the cleaned journey back into a string and update the DataFrame
        cleaned_data.at[index, target_column] = '-'.join(result_list)
    
    return cleaned_data
```

```python
'''
Joins all user journeys from different sessions into one user journey string. Has the option to select how many user sessions 
should be included in the new string

Parameters:
sessions = 'All' if all user sessions are required, else use integer to specify number of sessions
count_from= 'first' OR 'last' counts number of 'sessions' from either their first of last session

Returns:
The user ID with all of their page visits for the specified number of sessions
'''

def group_by(data, group_column='user_id', target_column='user_journey', sessions='All', count_from='last'):
    # Create a new DataFrame to store the grouped data
    grouped_data = pd.DataFrame()

    # Group data by 'group_column' (e.g., 'user_id')
    grouped = data.groupby(group_column)

    for group, group_df in grouped:
        if sessions == 'All':
            # If 'All' sessions are to be included, join all user journey strings
            grouped_user_journey = '-'.join(group_df[target_column])
        else: 
            # Sort the sessions by session id
            group_df = group_df.sort_values(by='session_id', ascending=True)
            
            if count_from == 'first':
                grouped_user_journey = '-'.join(group_df.head(sessions)[target_column])
            elif count_from == 'last':
                grouped_user_journey = '-'.join(group_df.tail(sessions)[target_column])
        subscription = group_df['subscription_type'].unique()
        subscription_string = ', '.join(subscription)
        
        # Create a new row for the grouped data
        grouped_row = pd.Series({group_column: group, 'subscription_type': subscription_string,  target_column: grouped_user_journey})
        grouped_data = grouped_data.append(grouped_row, ignore_index=True)

    return grouped_data
```
